```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ANAS NASIR.csv')
```

```python
head_insights = df.head(15)
head_insights
```

```python
df.fillna(0, inplace = True)
df.to_csv('ANAS NASIR.csv', index = False)
print("Missing values filled with 0 and saved to 'ANAS NASIR.csv'.")
df
```

```python
stack_df = df.stack()
stack_df
```

```python
index_reset = df.stack().reset_index()
index_reset.head(15)
```

```python
index_reset.columns = ['Bank_ID', 'Metric_Year', 'Financial_Value']
index_reset.head(15)
```

```python
melted_name_only = df.melt(id_vars=[df.columns[0]], var_name='Year', value_name='Value')
melted_name_only.rename(columns={'Unnamed: 0': 'Bank_ID'}, inplace=True)
melted_name_only.head(15)
```

```python
print("="*70)
print("SHAPE COMPARISON - Original vs Tidied Formats")
print("="*70)
print(f"Original Dataset:        {df.shape[0]} rows × {df.shape[1]} columns")
print(f"Stack + Reset_Index:     {index_reset.shape[0]} rows × {index_reset.shape[1]} columns")
print(f"Melt Method:             {melted_name_only.shape[0]} rows × {melted_name_only.shape[1]} columns")
print("="*70)
```

```python
# Create analyzable format: Metrics as columns, Years as rows
# Using the original melted data structure
analyzable_final = index_reset.copy()

# Convert to proper numeric format where possible
analyzable_final['Financial_Value'] = pd.to_numeric(analyzable_final['Financial_Value'], errors='coerce')

# Create pivot table: Rows = Metric_Year, Columns = Bank_ID, Values = Financial_Value
analyzable_pivoted = analyzable_final.pivot_table(
    index='Metric_Year', 
    columns='Bank_ID', 
    values='Financial_Value', 
    aggfunc='first'
)

print("="*70)
print("ANALYZABLE FORMAT - Metrics as Rows, Banks as Columns")
print("="*70)
print(f"Shape: {analyzable_pivoted.shape[0]} metrics/years × {analyzable_pivoted.shape[1]} banks")
print("\nFirst 10 rows:")
print(analyzable_pivoted.head(10))
```
