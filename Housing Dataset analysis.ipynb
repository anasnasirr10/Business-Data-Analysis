# **Feature Engineering**
FE is the process of creating, transforming ad selecting the most revelvant variables (features) from raw data to improve model perfromance.

**Features**: Relevant variables 

-----

```python
import pandas as pd
import numpy as np
```

```python
housing_csv = pd.read_csv('Housing.csv')
```

```python
housing_csv.head(15)
```

```python
housing_new = housing_csv.select_dtypes(include = np.number)
print(housing_new.head(15))
print(housing_new.shape)
print(housing_new.tail(15))
```

## **Absolute Maximum Scaling**



```python
max_abs = np.max(np.abs(housing_new), axis = 0)
print(max_abs)
scaled_df = housing_new / max_abs
print(scaled_df.head(15))
print(scaled_df.tail(15))
```

## **Min Max scaling**

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled_data = scaler.fit_transform(housing_new)
print(scaled_data)

scaled_df = pd.DataFrame(scaled_data, columns = housing_new.columns)

print(scaled_df.head(15))

print(scaled_df.tail(15))
```

## **Eucledian Method**

```python
from sklearn.preprocessing import Normalizer

scaler = Normalizer()

scaled_data = scaler.fit_transform(housing_new)

scaled_df = pd.DataFrame(scaled_data, columns = housing_new.columns)

print(scaled_df.head(15))
print(scaled_df.tail(15))
```

## **Standardization**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_data = scaler.fit_transform(housing_new)
print(scaled_data)
scaled_df = pd.DataFrame(scaled_data, columns = housing_new.columns)

print(scaled_df.head(15))
print(scaled_df.tail(15))
```

## **Robust Scaling**

Robust scaling uses median and Interquartile range instead of mean and and standard deviation making trasformation less sensitive to outliers, making transformation more robust with skewed distribution.

**IQR = Q3 - Q1**

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()

scaled_data = scaler.fit_transform(housing_new)

scaled_df = pd.DataFrame(scaled_data, columns = housing_new.columns)

print(scaled_df.head(15))
print(scaled_df.tail(15))
```
