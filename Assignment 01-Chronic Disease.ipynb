```python
def summary_stats(chronic_diseases):
     desc = chronic_diseases.describe().T
     desc['skewness'] = chronic_diseases.skew()
     desc['kurtosis'] = chronic_diseases.kurt()
     return desc
```

```python
import statsmodels.api as sm
```

```python
y = chronic_diseases['Chronic_kidney_disease']
x = chronic_diseases[['BP_Abnormality','Level_of_Hemoglobin','BMI','Smoking']]
```

```python
x = sm.add_constant(x)
```

```python
x_clean = x.dropna()
y_clean = y.loc[x_clean.index]
model = sm.OLS(y_clean, x_clean).fit()
```

```python
print(model.summary())
```
