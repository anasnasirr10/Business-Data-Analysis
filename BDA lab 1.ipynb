# **The Box-and-Whisker Plot**

The Box-and-Whisker plot is a powerful visual summary of distrubution. Let's interpret the output: 

- The **Box** is used to represent the median and the IQR (Interquartile Range). The bottom edge of the box is the 25th percentile and the top edge is the 75th percentile. The box is used to show the typical prices of cars in a dataset. The line inside the box is median and it is giving us information about that half of cars are priced below and half are priced above the median. The median is slightly close to the bottom of the box which suggests that the distrubution is right-skewed (more expensive cars pulling the average up). 

- Outliers are the extreme values that affect the normality of any dataset, they are the anomalies. In this diagram we can see extreme outliers that have moved nearly towards 200,000 explanaing to us that there are high-end vechiles that priced at a higher price value. 

- The box is compressed towards the bottom telling us that **MSRP** is right-skewed.

```python
def summary_stats(car):
     desc = car.describe().T
     desc['Skewness'] = car.skew()
     desc['Kurtosis'] = car.kurtosis()
     return desc
```

# **Multivariate Regression Analysis**
Multivariate regression analysis is a statistical technique that is used to estimate/predict the effect of relationship of one dependent variable with two or more independent variable.
Regression analysis is also is used to argue on the significance of the model to predict if there is any irrelevant variable in the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib as plt
np.random.seed(1234)
```

```python
import statsmodels.api as sm
```

```python
y = car['Invoice']
```

```python
x = car[['EngineSize', 'Cylinders', 'Horsepower', 'MPG_City']]
```

```python
x = sm.add_constant(x)
```

```python
# Drop rows with NaNs in x or y
x_clean = x.dropna()
y_clean = y.loc[x_clean.index]
model = sm.OLS(y_clean, x_clean).fit()
```

```python
print(model.summary())
```
