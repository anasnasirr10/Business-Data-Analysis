## <u>**Parametric and Non-parametric tests**</u>


## <u>**Parametric Test**</u>

Parametric test is a statistical technique that makes assumptions on basis of how much your data distribution is normal. It is suitable for quantitiatve (numeric) and continuous variables. They are numeric quantitities that represent whole population. 


## <u>**Non-Parametric Test**</u>

The test that doesn't rely on any statistical distribution; and hence the name suggets it is also known as "distribution-free" hypothesis test. These tests do not have parameters of the population. Non-parametric test is suitable for variables that are categorical in nature.

## **<u>Introduction to Hypothesis</u>**

A hypothesis is an assumption about a population that we test using statistical methods. It helps us decide whether the observed data supports or goes against them.

There are two types of hypothesis to claim whether our test supports or goes against it.

<u>**Null Hypothesis (Ho)**</u>

The hypothesis test that assumes that there is no effect or relation, or no difference. 

<u>**Alternate Hypothesis (Ha)**</u>

The hypothesis that assumes that a relationship, effect or difference exists.

<u>**Types of Parametic Tests**</u>

1. **t-test:** A specific type of test that is used for checking if there is a significant difference between means of the two groups concerned. It is categorized into two types:

     **i. One-sample t-test** is used to check if there is a significant difference a sample and hypothesized population means.

     **ii. Two-sample t-test** is used to measure significant difference between two groups.   


2. **ANOVA (Analysis of Variance)**

     ANOVA is a statistical inference test used for comparing multiple groups. It usually compares more than two datasets. It has variations like **One-way ANOVA** (Multile group comparison using single independent variable), **Two-way ANOVA** (multiple groups comparison based on two independet variables), and the last **N-way Anova** (Multiple groups comparison based on N independent variables).

**<u>Type of Non-Parametric Test</u>**

**Chi-square test**, a test that is determined by a significant difference or relationship between two categorical variables from single population. In general, this test assesses whether distribution of categorical variables differ from each other. It is also known as Chi-square goodness of fit test or test for independence.

## <u>**One-sample t-test**</u>

```python
import pandas as pd
from scipy import stats
```

```python
cars = pd.read_csv('cars.csv')
```

### **Targeted variables of One-sample t-test**

1. DriveTrain 
2. MSRP

**Hypothesis**: To examine whether the mean MSRP differs significantly between Front-wheel and Rear-wheel drive vehicles.

```python
print('\nTest 1: T-test on MSRP by DriveTrain (Front vs Rear)')

front = cars[cars['DriveTrain'] == 'Front']['MSRP']
rear = cars[cars['DriveTrain'] == 'Rear']['MSRP']
t_stat1, p_val1 = stats.ttest_ind(front, rear, equal_var = False)

print(f"T-statistic = {t_stat1:.3f}")
print(f"P-value = {p_val1: .5f}")
```

## **Decision**

```python
if p_val1 < 0.05:
    print("Reject the null hypothesis: There is a significant difference in mean MSRP between Front-wheel and Rear-wheel drive vehicles, alternate hypothesis will be accepted.")
else:
    print("Fail to reject the null hypothesis: No significant difference in mean MSRP between Front-wheel and Rear-wheel drive vehicles, null hypothesis will be accepted.")
```

## **<u>Two Sample T-test**</u>

## **Variables targeted**

1. MSRP
2. Origin

## **Hypothesis**

To evaluate whether cars manufactured in Asia and USA differ in their mean MSRP

```python
print('\n Test 2: T-test on MSRP by Origin (Asia vs USA)')

asia = cars[cars['Origin'] == 'Asia']['MSRP']
usa = cars[cars['Origin'] == 'USA']['MSRP']

t_stat2, p_val2 = stats.ttest_ind(asia, usa, equal_var = False)

print(f"T-statistic = {t_stat2:.3f}")
print(f"P-value = {p_val2: .5f}")

if p_val2 < 0.05:
    print("Reject the null hypothesis: There is a significant difference in mean MSRP between cars manufactured in Asia and USA, alternate hypothesis will be accepted.")
else:
     print("Fail to reject the null hypothesis: No significant difference in mean MSRP between cars manufactured in Asia and USA, null hypothesis will be accepted.")
```

## <u>**Two-way ANOVA**</u>

## **Hypothesis**

To determine whether **Origin** and DriveTrain individually and collectively influence the **MSRP** of cars.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

cars = cars.dropna(subset = ['Origin', 'MSRP', 'DriveTrain'])

model = ols('MSRP ~ C(Origin) + C(DriveTrain) + C(Origin):C(DriveTrain)', data = cars).fit()
anova_table = sm.stats.anova_lm(model, typ = 2)

print(anova_table)
```

```python
print('\nHo1: No difference in mean MSRP between cars manufactured in Asia, USA, and Europe.')
print('Ho2: No difference in mean MSRP between Front-wheel and Rear-wheel drive vehicles.')
print('Ho3: No interaction effect between Origin and DriveTrain on mean MSRP.')

print('\nDecision:')

for factor, p in anova_table['PR(>F)'].items():
    if p < 0.05:
        print(f"Reject the null hypothesis for {factor}: Significant effect on mean MSRP.")
    else:
        print(f"Fail to reject the null hypothesis for {factor}: No significant effect on mean MSRP.")
```

## <U>**N-Way ANOVA**</U>

## **Hypothesis**

To investigate how **Origin**, **DriveTrain**, and **Cylinders** collectively and interactibely affect car prices.

```python
cars = cars.dropna(subset = ['Origin', 'MSRP', 'DriveTrain', 'Cylinders'])

model = ols('MSRP ~ C(Origin) + C(DriveTrain) + C(Cylinders) + C(Origin):C(DriveTrain) + C(Origin):C(Cylinders) + C(DriveTrain):C(Cylinders) + C(Origin):C(DriveTrain):C(Cylinders)', data = cars).fit()
anova_table = sm.stats.anova_lm(model, typ = 2)

print(anova_table)

print('\nHypotheses:')

print('Ho1: None of the main or interaction effects significantly affect mean MSRP.')
print('Ho2: At least one main or interaction effect significantly affects mean MSRP.')

print('\nDecision:')

for factor, p in anova_table['PR(>F)'].items():
     if p < 0.05:
          print(f"Reject the null hypothesis for {factor}: Significant effect on mean MSRP.")
     else:
          print(f"Fail to reject the null hypothesis for {factor}: No significant effect on mean MSRP.")
```

## <u>**Non-Parametric Test**</u>

## **Chi-Sqare Test**

## **Hypothesis**

To determine if there is an association between the type of DriveTrain (Front, Rear, All) and the Origin of the car (Asia, USA, Europe).

```python
from scipy.stats import chi2_contingency

cars = cars.dropna(subset = ['Origin', 'DriveTrain'])

contingency_table = pd.crosstab(cars['Origin'], cars['DriveTrain'])

print("Contingency Table")

chi2, p_val, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square statistic = {chi2:.3f}")
print(f"P-value = {p_val:.5f}")
print(f'Degrees of Freedom = {dof}')
print(F"P-value = {p_val:.5f}")

pd.DataFrame(expected, index = contingency_table.index, columns = contingency_table.columns)
```

```python
print('\nHo: Origin and DriveTrain are independent.')
print('Ha: Origin and DriveTrain are not independent.')

print('\nDecision:')

if p_val < 0.05:
     print("Reject the null hypothesis: There is a significant association between Origin and DriveTrain.")
else:
     print("Fail to reject the null hypothesis: No significant association between Origin and DriveTrain.")
     
```
