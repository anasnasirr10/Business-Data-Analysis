```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Your data matrix X
X = np.array([
    [2.5, 2.4, 1.2],
    [0.5, 0.7, 0.3],
    [2.2, 2.9, 1.6],
    [1.9, 2.2, 1.0],
    [3.1, 3.0, 1.8]
])

# Apply PCA
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)

# Plot original 3D data
fig = plt.figure(figsize=(15, 5))

ax1 = fig.add_subplot(131, projection='3d')
ax1.scatter(X[:, 0], X[:, 1], X[:, 2], color='blue')
ax1.set_title('Original 3D Data')
ax1.set_xlabel('X1')
ax1.set_ylabel('X2')
ax1.set_zlabel('X3')

# Plot 2D PCA projection (first two principal components)
ax2 = fig.add_subplot(132)
ax2.scatter(X_pca[:, 0], X_pca[:, 1], color='green')
ax2.set_title('2D PCA Projection (PC1 vs PC2)')
ax2.set_xlabel('Principal Component 1')
ax2.set_ylabel('Principal Component 2')
ax2.grid(True)

# Plot 1D PCA projection (first principal component only)
ax3 = fig.add_subplot(133)
ax3.scatter(X_pca[:, 0], np.zeros_like(X_pca[:, 0]), color='red')
ax3.set_title('1D PCA Projection (PC1)')
ax3.set_xlabel('Principal Component 1')
ax3.set_yticks([])
ax3.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D  # Needed for 3D plot

# 4D dataset
X = np.array([
    [2.5, 2.4, 1.2, 3.1],
    [0.5, 0.7, 0.3, 0.8],
    [2.2, 2.9, 1.6, 2.9],
    [1.9, 2.2, 1.0, 2.5],
    [3.1, 3.0, 1.8, 3.3],
    [2.3, 2.7, 1.5, 2.8]
])

# -------- 4D Representation --------
fig = plt.figure(figsize=(12, 5))

ax = fig.add_subplot(121, projection='3d')
sc = ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=X[:, 3], cmap='viridis', s=60)
ax.set_title('4D Representation (Color = 4th Dim)')
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
cb = plt.colorbar(sc, ax=ax)
cb.set_label('Feature 4')

# -------- PCA: Reduce to 2D --------
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# 2D PCA Projection
ax2 = fig.add_subplot(122)
ax2.scatter(X_pca[:, 0], X_pca[:, 1], color='darkorange')
for i, (x, y) in enumerate(X_pca):
    ax2.text(x, y, f"P{i+1}", fontsize=8, ha='right')
ax2.set_title('2D PCA Projection')
ax2.set_xlabel('Principal Component 1')
ax2.set_ylabel('Principal Component 2')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D

# Step 0: Generate synthetic 4D data (20 entries)
np.random.seed(42)
X = np.random.rand(20, 4) * 10
print("Original Data (X):\n", X)

# Step 0.1: 4D Plot using 3D + color for 4th feature
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

x = X[:, 0]  # Feature 1
y = X[:, 1]  # Feature 2
z = X[:, 2]  # Feature 3
color = X[:, 3]  # Feature 4 (mapped to color)

scatter = ax.scatter(x, y, z, c=color, cmap='viridis', edgecolor='k', s=70)
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
plt.title('Original 4D Data (Color = Feature 4)')
cbar = plt.colorbar(scatter, ax=ax, pad=0.1)
cbar.set_label('Feature 4')
plt.show()

# Step 1: Standardize the data
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
#print("\nStep 1: Standardized Data:\n", X_std)

# Step 2: Covariance matrix
cov_matrix = np.cov(X_std.T)
#print("\nStep 2: Covariance Matrix:\n", cov_matrix)

# Step 3: Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
#print("\nStep 3: Eigenvalues:\n", eigenvalues)
#print("Eigenvectors:\n", eigenvectors)

# Step 4: Sort eigenvectors by decreasing eigenvalues
sorted_idx = np.argsort(eigenvalues)[::-1]
eigenvalues_sorted = eigenvalues[sorted_idx]
eigenvectors_sorted = eigenvectors[:, sorted_idx]
#print("\nStep 4: Sorted Eigenvalues:\n", eigenvalues_sorted)
#print("Sorted Eigenvectors:\n", eigenvectors_sorted)

# Step 5: Project data onto top 2 principal components
W = eigenvectors_sorted[:, :2]
X_pca = X_std.dot(W)
#print("\nStep 5: Projected Data (PCA result):\n", X_pca)

# Plot PCA projection
plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c='blue', edgecolor='k')
plt.title('PCA Projection (PC1 vs PC2)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D

# Generate 100 entries with 3 features
np.random.seed(42)
X = np.random.rand(100, 3) * 10

# Standardize the data
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# PCA computation
cov_matrix = np.cov(X_std.T)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
sorted_idx = np.argsort(eigenvalues)[::-1]
W = eigenvectors[:, sorted_idx[:2]]
X_pca = X_std @ W

# Plotting
fig = plt.figure(figsize=(14, 5))

# Original 3D data
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.scatter(X[:, 0], X[:, 1], X[:, 2], c='green', edgecolor='k')
ax1.set_title('Original 3D Data')
ax1.set_xlabel('Feature 1')
ax1.set_ylabel('Feature 2')
ax1.set_zlabel('Feature 3')

# PCA 2D projection
ax2 = fig.add_subplot(1, 2, 2)
ax2.scatter(X_pca[:, 0], X_pca[:, 1], c='blue', edgecolor='k')
ax2.set_title('PCA Projection (PC1 vs PC2)')
ax2.set_xlabel('PC1')
ax2.set_ylabel('PC2')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Generate synthetic data: 100 entries, 10 features
np.random.seed(42)
X = np.random.rand(100, 10) * 10

# Standardize
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# PCA computation
cov_matrix = np.cov(X_std.T)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
sorted_idx = np.argsort(eigenvalues)[::-1]
W = eigenvectors[:, sorted_idx[:2]]
X_pca = X_std @ W

# Prepare plots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: Pairplot of first 3 original features
df_partial = pd.DataFrame(X[:, :3], columns=['Feature 1', 'Feature 2', 'Feature 3'])
sns.scatterplot(data=df_partial, x='Feature 1', y='Feature 2', ax=axes[0], edgecolor='k')
axes[0].set_title('Original Data (Feature 1 vs Feature 2)')

# Right: PCA projection
axes[1].scatter(X_pca[:, 0], X_pca[:, 1], c='blue', edgecolor='k')
axes[1].set_title('PCA Projection (PC1 vs PC2)')
axes[1].set_xlabel('PC1')
axes[1].set_ylabel('PC2')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
