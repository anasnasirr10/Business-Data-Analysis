```python
reset_indexed = df.reset_index()
reset_indexed.head(15) 
```

Each row represents a different financial metric of Bank AL- Habib Ltd, including items such as **Loans Net**, **Total Assets**, **Deposits**, and **Shareholder's Equity**, with corresponding numeric numerical values listed for each year from 1996 to 2007. 
The data is structured so that each finacial variable appears as a seperate row containing a full time series of annual data. One of the rows contains corrupted text instead of numeric values, indicating that the dataset includes at least one improperly imported or malformed entry that will require cleaning.

```python
reset_indexed.rename(columns={'Unnamed: 0': 'Bank_ID'}, inplace=True)
print("Column renamed: 'Unnamed: 0' → 'Bank_ID'")

print("\nDataframe columns:")
print(reset_indexed.columns.tolist())

print("\nFirst 10 rows with renamed column:")
reset_indexed.head(10)
```

The dataframe shows financial data for banks in a wide panel format. Each row represent a specific financial metric for one bank (identified by Bank_ID). The columns labeled 1996-2010 show the year-wise values of that metric. Teh metrics include items such as **Loans_Net**, **Prov_Loss**, **Total_Assets**, **Non-performing_loans**, **%Tot loans** and **Shareholders_equity**.

```python
bank_split = reset_indexed['Name'].str.split(pat=' - ', expand=True)

bank_split.insert(0, 'Bank_ID', reset_indexed['Bank_ID'].values)

bank_split.columns = ['Bank_ID', 'Bank_Name', 'Metric', 'Additional']

print("Split 'Name' column with Bank_ID:")
print(f"Shape: {bank_split.shape}")
print("\nFirst 15 rows:")
print(bank_split.head(15))
```

```python
id_vars = ['Bank_ID', 'Name', 'CURRENCY']

year_columns = [col for col in reset_indexed.columns if col not in id_vars and col != 'index']

# Melt the dataset
melted_full = reset_indexed.melt(
    id_vars=id_vars,
    value_vars=year_columns,
    var_name='Year',
    value_name='Value'
)

print("="*70)
print("Melted dataset - Wide to Long Format")
print("="*70)
print(f"Original shape: {reset_indexed.shape}")
print(f"Melted shape: {melted_full.shape}")
print(f"\nColumns: {melted_full.columns.tolist()}")
print(f"\nFirst 20 rows of melted data:")
print(melted_full.head(20))
```

The output indicates that the column was split using an incorrect pattern, so Pandas seperated the text mainly by spaces. Rows containing full bank names and metrics were split reasonably well, but rows were misalligned, leaving empty metric fields. 
Overall, the split worked inconsistenly because the pattern didn't match the actual data structure.
